```python
#No1
import pandas as pd
m_store=pd.read_csv("/content/drive/MyDrive/100Knock/chap1/m_store.csv")
m_store
```

```python
m_area=pd.read_csv("/content/drive/MyDrive/100Knock/chap1/m_area.csv")
m_area
```

```python
tbl_order_4=pd.read_csv("/content/drive/MyDrive/100Knock/chap1/tbl_order_202004.csv")
tbl_order_4
```

```python
#No2
tbl_order_5=pd.read_csv("/content/drive/MyDrive/100Knock/chap1/tbl_order_202005.csv")
tbl_order_5
tbl_order_6=pd.read_csv("/content/drive/MyDrive/100Knock/chap1/tbl_order_202006.csv")
tbl_order_6
```

```python
order_all = pd.concat([tbl_order_4,tbl_order_5],ignore_index=True)
order_all = pd.concat([order_all,tbl_order_6],ignore_index=True)
order_all
```

```python
#結合の件数確認
len(order_all) == len(tbl_order_4)+len(tbl_order_5)+len(tbl_order_6)
```

```python
#5
order_all.isnull().sum()
tbl_order_6.isnull().sum()

order_all[order_all['total_amount'].isnull()]
order_all['total_amount'].fillna(order_all['total_amount'].mean())      
order_all['status'].fillna(2)  
```

```python
order_all.describe()
```

```python
order_data=order_all.loc[order_all['store_id']!=999]
order_data
```

```python
#7
order_data=pd.merge(order_data,m_store,on='store_id',how='left')
order_data
```

```python
order_data.rename(columns={'store_name_x': 'store_name','area_cd_x':'area_cd'},inplace=True)
```

```python
order_data=pd.merge(order_data,m_area,on='area_cd',how='left')
order_data
```

```python
order_data.drop(columns=['store_name_y','area_cd_y'],axis=1,inplace=True)
order_data.head()
```

```python
#8
order_data.loc[order_data['takeout_flag']==0,'takeout_name']='デリバリー'
order_data.loc[order_data['takeout_flag']==1,'takeout_name']='お持ち帰り'

order_data.loc[order_data['status']==0,'status_name']='受付'
order_data.loc[order_data['status']==1,'status_name']='お支払い済'
order_data.loc[order_data['status']==2,'status_name']='お渡し済'
order_data.loc[order_data['status']==9,'status_name']='キャンセル'

order_data.head()
```

```python
#11
order_data = order_data.loc[(order_data['status'] == 1)|(order_data['status'] == 2)]
analyze_data = order_data[['store_id', 'customer_id', 'coupon_cd',
  'order_accept_date', 'delivered_date', 'total_amount',
  'store_name', 'wide_area', 'narrow_area',
  'takeout_name', 'status_name']]
analyze_data.head()
```
